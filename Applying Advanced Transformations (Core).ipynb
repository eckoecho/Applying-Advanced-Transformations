```python
#Clean the files and combine them into one final DataFrame.
import pandas as pd
import json
```

```python
#Importing the info csv and converting to df
info = pd.read_csv("Data/superhero_info - superhero_info.csv")
info.head()
```

```python
#Importing the info csv and converting to df
powers = pd.read_csv("Data/superhero_powers - superhero_powers.csv")
powers.head()
```

```python
## adding expand=True
info[["Hero","Publisher"]] = info["Hero|Publisher"].str.split("|", expand=True)
info
```

```python
#Dropping column we split into 2
info.drop(columns="Hero|Publisher", inplace=True)
info
```

# Cleaning Measurement

```python
## examining a single value from the measurements col 
# 0 indicates the row
specs = info.loc[0, "Measurements"]
print(type(specs))
```

```python
specs
```

```python
## use .str.replace to replace all single quotes with double quotes
info["Measurements"] = info["Measurements"].str.replace("'", '"')
## use .str.replace to replace all single quotes with double quotes
info["Measurements"] = info["Measurements"].str.replace("'", '"')
## Apply the json.loads to the full column
info["Measurements"] = info["Measurements"].apply(json.loads)
```

```python
info["Measurements"].head()
```

```python
#use the pandas pd.Series function on our column of dictionaries to turn into a new column
measurements = info["Measurements"].apply(pd.Series)
measurements
```

```python
#Removing cm from height column, kg from weight
measurements["Height"] = measurements["Height"].str.replace(" cm","")
measurements["Weight"] = measurements["Weight"].str.replace(" kg","")
measurements[["Height", "Weight"]]
```

```python
#Converting string to float, then float to int
measurements[["Height", "Weight"]] = measurements[["Height", "Weight"]].astype("float")
measurements[["Height", "Weight"]] = measurements[["Height", "Weight"]].astype("int64")
```

```python
# concat measurements with original dataframe
info = pd.concat((info, measurements), axis=1)
info.head()
```

```python
#Drop measurements column
info = info.drop(columns = "Measurements")
#Ensuring Height and Weight are integers
info.info()
```
